```python
import os

from ultralytics import YOLO
from ultralytics.utils.plotting import plot_results
from PIL import Image
```

```python
os.environ['YOLO_DATASETS_DIR'] = './datasets'
```

# PRETRAINING

```python
# Load a model
model = YOLO('yolov8s.pt')  # load a pretrained model (recommended for training)

# Train the model
results = model.train(data='VisDrone.yaml', 
                      epochs=100, 
                      imgsz=512, 
                      name="VisDrone_S",
                      exist_ok=True,
                     )
```

# TRAINING

```python
# Load a model
model = YOLO('runs/detect/VisDrone_S/weights/best.pt')  # load a pretrained model on VisDrone dataset

# Train the model
results = model.train(data='birdDetection3_yolo8.yaml', 
                      epochs=300, 
                      imgsz=512, 
                      batch=16, 
                      name="birdDetection_yolo8",
                      exist_ok=True,
                      )
```

# MODEL

```python
model = YOLO('runs/detect/birdDetection_yolo8/weights/best.pt') 
```

```python
metrics = model.val()
```

# RESULTS
